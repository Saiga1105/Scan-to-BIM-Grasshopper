```python
dmap = dmap2
shape = np.shape(dmap)
print("Dimensions: ", shape)

heigth = shape[0]
print("heigth: ", heigth)
width = shape[1]
print("width: ", width)
nmbr_pts = width * heigth
i = 0
pts = np.empty([nmbr_pts, 3])

h = 0

while h < heigth:
    w = 0
    while w < width:
        pts[i] = [h,w,dmap[h,w]/255*1000]
        i = i+1
        w = w+1
    h = h+1
        

print("PCD Dimensions: ", np.shape(pts))
print(pts)

```

```python
#Visualization
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(pts)
o3d.visualization.draw_geometries([pcd])
```
